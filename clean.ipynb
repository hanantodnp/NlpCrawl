```python
import pandas as pd
import re
```

```python
data = pd.read_csv('data/hanan.csv')
data
```

```python
def cleansing(sent):
    string = sent.lower()
    string = re.sub(r'http\S+', '', string)  # Remove URLs
    string = re.sub(r'@\w+', '', string)     # Remove mentions
    string = re.sub(r'#\w+', '', string)     # Remove hashtags
    string = re.sub(r'\d+', '', string)      # Remove angka
    string = re.sub(r'[^\w\s]', '', string)  # Remove punctuation
    return string
```

```python
data['full_text'] = data.full_text.apply(cleansing)
data.head()
```

```python
data = data[['full_text']]
data.head
```

```python
data.to_csv('data/clean_hanan.csv', index=False)
```
